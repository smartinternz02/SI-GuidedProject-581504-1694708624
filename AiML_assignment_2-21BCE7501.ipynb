# SCATTER PLOT

```python
sns.scatterplot(x="alcohol", y="total", data=dset)
plt.title("Scatterplot: Alcohol vs Total Crashes")
plt.xlabel("Alcohol Consumption")
plt.ylabel("Total Crashes")
```

Inference: Positive correlation between alcohol consumption and total crashes from the above plot, i.e as the alcohol consumption increases the total crashes increases.

```python
sns.scatterplot(x="speeding", y="total", data=dset)
plt.title("Scatterplot: Speeding vs Total Crashes")
plt.xlabel("Speeding")
plt.ylabel("Total Crashes")
```

Inference: Speeding doesn't show a clear linear trend with total crashes.

# LINE PLOT

```python
sns.lineplot(x="alcohol", y="total", data=dset)
plt.title("Lineplot: Alcohol vs Total Crashes")
plt.xlabel("Alcohol Consumption")
plt.ylabel("Total Crashes")
```

Inference: No obvious linear trend in the relationship between alcohol consumption and total crashes.


```python
sns.lineplot(x="speeding", y="total", data=dset)
plt.title("Lineplot: Speeding vs Total Crashes")
plt.xlabel("Speeding")
plt.ylabel("Total Crashes")
```

Inference: Speeding doesn't exhibit a consistent linear relationship with total crashes.

# DISTRIBUTION PLOT

```python
sns.histplot(dset["not_distracted"], kde=True)
plt.title("Distplot: Not Distracted")
plt.xlabel("Not Distracted")
```

Inference: The distribution of "not_distracted" values is right-skewed

```python
sns.histplot(dset["alcohol"], kde=True)
plt.title("Distplot: Alcohol Consumption")
plt.xlabel("Alcohol Consumption")
```

Inference: The distribution of alcohol consumption appears to be right-skewed as well

#BOX PLOT

```python
sns.boxplot(x="alcohol", y="total", data=dset)
plt.title("Boxplot: Alcohol vs Total Crashes")
plt.xlabel("Alcohol Consumption")
plt.ylabel("Total Crashes")
```

Inference: The boxplot shows the distribution of total crashes for different levels of alcohol consumption. The lines indicates the outliers

```python
sns.boxplot(x="speeding", y="total", data=dset)
plt.title("Boxplot: Speeding vs Total Crashes")
plt.xlabel("Speeding")
plt.ylabel("Total Crashes")
```

Inference: The boxplot illustrates the distribution of total crashes for different levels of speeding. The lines indicate the outliers.

# BAR PLOT

```python
sns.barplot(x="alcohol", y="total", data=dset)
plt.title("Barplot: Alcohol vs Total Crashes")
plt.xlabel("Alcohol Consumption")
plt.ylabel("Total Crashes")
```

Inference: The barplot displays the mean total crashes for different levels of alcohol consumption. So, if the alcohol consumption is high, then total crashes are also high.

```python
sns.barplot(x="speeding", y="total", data=dset)
plt.title("Barplot: Speeding vs Total Crashes")
plt.xlabel("Speeding")
plt.ylabel("Total Crashes")
```

Inference: The barplot shows the mean total crashes for different levels of speeding. The crashes are high even at low speed levels also.

# JOINT PLOT

```python
sns.jointplot(x="alcohol", y="total", data=dset, kind="scatter")
plt.suptitle("Jointplot: Alcohol vs Total Crashes", y=1.02)
```

Inference: The plot in the jointplot reveals the relationship between alcohol consumption and total crashes. So, as the alcohol increases, the total crashes also increase.

```python
sns.jointplot(x="speeding", y="total", data=dset, kind="scatter")
plt.suptitle("Jointplot: Speeding vs Total Crashes", y=1.02)
```

Inference: The plot in the jointplot shows the relationship between speeding and total crashes. The plot is not in a specific pattern.

```python
sns.relplot(x="alcohol", y="total", data=dset, kind="scatter")
plt.title("Relationalplot: Alcohol vs Total Crashes")
plt.xlabel("Alcohol Consumption")
plt.ylabel("Total Crashes")
```

Inference: The plot in the relationalplot visualizes the relationship between alcohol consumption and total crashes and it is directly proportional.

# RELATION PLOT

```python
sns.relplot(x="speeding", y="total", data=dset, kind="scatter")
plt.title("Relationalplot: Speeding vs Total Crashes")
plt.xlabel("Speeding")
plt.ylabel("Total Crashes")
```

Inference: The scatter plot in the relationalplot illustrates the relationship between speeding and total crashes and it is not in a specific pattern.

```python
sns.countplot(x="alcohol", data=dset)
plt.title("Countplot: Alcohol Consumption")
plt.xlabel("Alcohol Consumption")
```

Inference: This countplot shows the frequency of different levels of alcohol consumption in the dataset and the count is maximum as 2 at a particular alcohol consumption and 1 otherwise.

# COUNT PLOT

```python
sns.countplot(x="speeding", data=dset)
plt.title("Countplot: Speeding")
plt.xlabel("Speeding")
```

Inference: This countplot displays the frequency of speeding incidents in the dataset and the count is maximum as 2 at a particular alcohol consumption and 1 otherwise.
